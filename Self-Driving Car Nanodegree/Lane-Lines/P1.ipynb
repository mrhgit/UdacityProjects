## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.
def my_lane_vertices(img,scale_factor=1.0):
    # Assume a 1.77777:1 width:height ratio, that is 960:540 or 1280:720
    [im_height, im_width, im_chan] = img.shape
    # Vertices are entered as (x,y), which (0,0) being the top-left of the image
    # Do not need to close the polygon by repeating the first entry
    # This shape is an hourglass shape
    bottom_y = im_height-1
    right_x = im_width - 1
    #top_y = bottom_y - 210*scale_factor
    #top_right = im_width/2 + 70*scale_factor
    #top_left = im_width/2 - 70*scale_factor
    top_y = bottom_y - 200*scale_factor
    top_right = im_width/2 + 60*scale_factor
    top_left = im_width/2 - 60*scale_factor
    #top_y = bottom_y - 190*scale_factor
    #top_right = im_width/2 + 50*scale_factor
    #top_left = im_width/2 - 50*scale_factor
    
    vertices = np.array([[(0,bottom_y),
                      (right_x, bottom_y),
                      (top_right,top_y),
                      (top_left,top_y)]], dtype=np.int32)
        
    return vertices

def find_line_params(line):
    [x1,y1,x2,y2] = line
    if np.abs(x2-x1) < 0.01:
        return None,None
    
    slope = (y2-y1)/(x2-x1*1.)
    bias = y1 - slope*x1 # y = mx + b ... b = (y-mx)
    return slope, bias

def find_x_for_y(line, y):
    [x1,y1,x2,y2] = line
    slope,bias = find_line_params(line)
    # y = mx + b ... x = (y-b)/m
    if slope is None:
        return None
    if slope==0:
        return int(x1+x2)/2
    return int((y - bias)/slope)

def find_line_length(line):
    [x1,y1,x2,y2] = line
    length = np.sqrt(np.power(x2-x1,2)+np.power(y2-y1,2))
    return length

def find_line_height(line):
    [x1,y1,x2,y2] = line
    return abs(y2 - y1)

def vertical_range_of_mask(vertices):
    min_y = 539
    max_y = 0
    for blah in vertices:
        for [x,y] in blah:
            if y < min_y: min_y = y
            if y > max_y: max_y = y
    return abs(max_y - min_y)

def get_line_segments(line, num_segments):
    [x1,y1,x2,y2] = line
    y_step = (y2-y1)/(num_segments*2 - 1)
    lines = []
    y = y1
    for i in range(0,num_segments*2,2):
        lines.append([find_x_for_y(line, int(y)), int(y), find_x_for_y(line, int(y+y_step)), int(y+y_step)])
        y += 2*y_step
    return lines


def my_draw_lines(img, lines, color=[255, 0, 0], thickness=2, max_line_vert=0):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """    
    [im_height, im_width, im_chan] = img.shape
    top_y = im_height - max_line_vert - 1
    bottom_y = im_height - 1

    # First, split into left and right lane lines
    left_lane_lines = []
    left_max_length = 0
    left_avg_top_x = 0
    left_avg_bottom_x = 0
    left_weight_sum = 0
    right_lane_lines = []
    right_max_length = 0
    right_avg_top_x = 0
    right_avg_bottom_x = 0
    right_weight_sum = 0

    
    # Decide which ones to keep and keep tabs on averages and maxes
    for [line] in lines:
        [x1,y1,x2,y2] = line
        slope,bias = find_line_params(line)
        if slope is None or abs(slope)>1 or abs(slope)<0.5:
            continue
        if slope > 0:
            if bias < -im_height:
                continue
            if x1 < im_width*.5 or x2 < im_width*.5:
                continue
            right_lane_lines.append(line)
            right_max_length = max(right_max_length,find_line_height(line))
            weight = np.power(find_line_length(line),2)
            right_avg_top_x += find_x_for_y(line,top_y)*weight
            right_avg_bottom_x += find_x_for_y(line,bottom_y)*weight
            right_weight_sum += weight
        else:
            if bias < im_height:
                continue
            if x1 > im_width*.5 or x2 > im_width*.5:
                continue
            left_lane_lines.append(line)
            left_max_length = max(left_max_length,find_line_height(line))
            weight = np.power(find_line_length(line),2)
            left_avg_top_x += find_x_for_y(line,top_y)*weight
            left_avg_bottom_x += find_x_for_y(line,bottom_y)*weight
            left_weight_sum += weight
    if len(left_lane_lines) > 0:
        left_avg_top_x = int(left_avg_top_x / left_weight_sum)
        left_avg_bottom_x = int(left_avg_bottom_x / left_weight_sum)
        if (left_max_length < 0.5 * max_line_vert):
            #left_lane_type = ["BROKEN",[0,0,255]]
            left_lane_type = ["SOLID",[0,255,0]]
            keep_alpha_left = 0.8
        else:
            left_lane_type = ["SOLID",[0,0,255]]
            keep_alpha_left = 0.2
    if len(right_lane_lines) > 0:
        right_avg_top_x = int(right_avg_top_x / right_weight_sum)
        right_avg_bottom_x = int(right_avg_bottom_x / right_weight_sum)
        if (right_max_length < 0.5 * max_line_vert):
            #right_lane_type = ["BROKEN",[255,0,0]]
            right_lane_type = ["SOLID",[0,255,0]]
            keep_alpha_right = 0.8
        else:
            right_lane_type = ["SOLID",[255,0,0]]
            keep_alpha_right = 0.2
    
    bg_border = 2
    gray_wash = 20
    num_dashes = 5
    
    # draw left line
    if len(left_lane_lines) == 0 and my_draw_lines.left_history is not None:
        left_lane = my_draw_lines.left_history['lane']
        left_lane_type = my_draw_lines.left_history['lane_type']
        left_bgcolor = [255,255,255]
        keep_alpha_left = 1.
    else:
        if len(left_lane_lines) > 0:
            left_bgcolor = left_lane_type[1]

    if len(left_lane_lines) > 0 or my_draw_lines.left_history is not None:
        left_lane = [left_avg_top_x, top_y, left_avg_bottom_x, bottom_y]
        if my_draw_lines.left_history is not None:
            left_lane = np.add(np.multiply(my_draw_lines.left_history['lane'], keep_alpha_left),np.multiply(left_lane,1-keep_alpha_left)).astype(np.uint32)
        if len(left_lane_lines) > 0:
            my_draw_lines.left_history = {'lane':left_lane,'lane_type':left_lane_type}
        if left_lane_type[0]=="SOLID":
            cv2.line(img, (left_lane[0], left_lane[1]), (left_lane[2], left_lane[3]), left_bgcolor, thickness+bg_border)
            cv2.line(img, (left_lane[0], left_lane[1]), (left_lane[2], left_lane[3]), left_lane_type[1], thickness)
        else:
            broken_lines = get_line_segments(left_lane, num_dashes)
            for [x1,y1,x2,y2] in broken_lines:
                cv2.line(img, (x1, y1), (x2, y2), left_bgcolor, thickness+bg_border)
                cv2.line(img, (x1, y1), (x2, y2), left_lane_type[1], thickness)            

    # draw right line
    if len(right_lane_lines) == 0 and my_draw_lines.right_history is not None:
        right_lane = my_draw_lines.right_history['lane']
        right_lane_type = my_draw_lines.right_history['lane_type']
        right_bgcolor = [255,255,255]
        keep_alpha_right = 1.
    else:
        if len(right_lane_lines) > 0:
            right_bgcolor = right_lane_type[1]

    if len(right_lane_lines) > 0 or my_draw_lines.right_history is not None:
        right_lane = [right_avg_top_x, top_y, right_avg_bottom_x, bottom_y]
        if my_draw_lines.right_history is not None:
            right_lane = np.add(np.multiply(my_draw_lines.right_history['lane'], keep_alpha_right),np.multiply(right_lane,1-keep_alpha_right)).astype(np.uint32)
        if len(right_lane_lines) > 0:
            my_draw_lines.right_history = {'lane':right_lane,'lane_type':right_lane_type}
        if right_lane_type[0]=="SOLID":
            cv2.line(img, (right_lane[0], right_lane[1]), (right_lane[2], right_lane[3]), right_bgcolor, thickness+bg_border)
            cv2.line(img, (right_lane[0], right_lane[1]), (right_lane[2], right_lane[3]), right_lane_type[1], thickness)
        else:
            broken_lines = get_line_segments(right_lane, num_dashes)
            for [x1,y1,x2,y2] in broken_lines:
                cv2.line(img, (x1, y1), (x2, y2), right_bgcolor, thickness+bg_border)
                cv2.line(img, (x1, y1), (x2, y2), right_lane_type[1], thickness)
                            
    return img

def reset_lane_smoothing():
    my_draw_lines.left_history = None
    my_draw_lines.right_history = None

def my_hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap, max_line_vert):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)

    line_img = my_draw_lines(line_img, lines, thickness=8, max_line_vert=max_line_vert)
    return line_img

def laneline_finder(img,reset_smoothing=False,show_images=False):
    if reset_smoothing:
        reset_lane_smoothing()
    [im_height, im_width, im_chan] = img.shape
    scale_factor = im_height / 540

    g_img = grayscale(img) # Convert to Grayscale
    #c_img = canny(g_img,low_threshold=50,high_threshold=255) # Edge Detection
    c_img = canny(img,low_threshold=50,high_threshold=255) # Edge Detection
    vertices = my_lane_vertices(img,scale_factor)  # Restrict Region of Interest
    r_img = region_of_interest(c_img,vertices)
    max_line_vert = vertical_range_of_mask(vertices)
    h_img = my_hough_lines(r_img, 1, np.pi/180.0, 35, min_line_len=10*scale_factor, max_line_gap=30*scale_factor, max_line_vert=max_line_vert) # Hough Transform
    h2_img = hough_lines(r_img, 1, np.pi/180.0, 35, min_line_len=10*scale_factor, max_line_gap=30*scale_factor) # Hough Transform
    w_img = weighted_img(h_img, img, =1., =0.5, =0.)

    if show_images:
        plt.figure(); plt.imshow(g_img,cmap='gray'); plt.title('Grayscale')
        plt.figure(); plt.imshow(c_img,cmap='gray'); plt.title('Canny')
        plt.figure(); plt.imshow(r_img,cmap='gray'); plt.title('Masked Region')
        plt.figure(); plt.imshow(h2_img,cmap='gray'); plt.title('Hough Lines')
        plt.figure(); plt.imshow(h_img); plt.title('Rendered Lane Lines')
    #return cv2.cvtColor(r_img, cv2.COLOR_GRAY2RGB)
    #return h_img
    #return h2_img
    return w_img
````
